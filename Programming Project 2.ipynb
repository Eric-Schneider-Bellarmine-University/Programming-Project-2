```python
import random

def mainframe(current_balance=1000, tickets_bought=0):
    '''
    The main sytem for the program, comprised of a while loop which 
    runs each round until the player chooses to quit or runs out of 
    money. The final culminating results of all the rounds is then 
    displayed.
    '''
    while current_balance//2:
        number_of_tickets,current_balance=get_wager(current_balance)
        tickets_bought+=number_of_tickets
        if number_of_tickets==0:
            break
        tickets=[]
        while len(tickets)<number_of_tickets:
            ticket=get_ticket()
            tickets.append(ticket) if ticket not in tickets else None
        winner=get_ticket()
        for chances in tickets:
            matches,powerball=check_ticket(chances,winner)
            current_balance+=get_winnings(matches,powerball)
            results[(matches,powerball)][1]+=1
    if not current_balance//2:
        print('You are out of money.') 
    else:
        print('You have chosen to quit the lottery')
    display_results(tickets_bought,current_balance)

def get_wager(current_balance):
    '''
    Takes the current balance and asks the player how may tickets 
    they want to buy for $2 a piece. If the amount is more than the
    player can afford, they will not be able to purchase. The game
    ends when a 0 is entered.
    '''
    while True:
        wager=input(f'You currently have ${current_balance}. Each lottery ticket costs $2. How many tickets would you like to purchase?\n')
        try: 
            int(wager)
        except:
            print('\nPlease only enter a valid currency.\n')
        else:
            wager=int(wager)
            if 2*wager>current_balance:
                print('\nWe do not accept debt in this game. Please refrain from betting money you do not have.\n')
            else:
                return wager,current_balance-2*wager

def get_ticket():
    '''
    A program which randomly generates the tickets in a list format
    '''
    return [random.randrange(1,70) for numbers in range(5)]+[random.randrange(1,27)]

def check_ticket(current_ticket,winner,matches=0):
    '''
    A program which takes the given ticket and compares it to the winning ticket, counting matches and isolating the powerball.
    '''
    powerball=True if current_ticket[-1]==winner[-1] else False
    for numbers in range(len(current_ticket)):
        if numbers in winner[:-1]:
            matches+=1
            winner.pop(winner.index(numbers))
        return matches,powerball

def get_winnings(matches,powerball):
    '''
    Takes the amount of mathces and whether or not the powerball 
    matches and displays a message for any winning tickets and 
    returning the designated amount of winnings to be added back 
    into the player's bank.
    '''
    if matches==5:
        if powerball:
            print('Congratulations, you have won the jackpot. $500,000,000 has been added to your balance.')
            return 500000000
        else:
            print('Congratulations, you matched 5 basic numbers. $1,000,000 has been added to your balance.')
            return 1000000
    elif matches==4:
        if powerball:
            print('Congratulations, you matched 4 basic numbers and the powerball. $50,000 has been added to your balance.')
            return 50000
        else:
            print('Congratulations, you matched 4 basic numbers. $100 has been added to your balance.')
            return 100
    elif matches==3:
        if powerball:
            print('Congratulations, you matched 3 basic numbers and the powerball. $100 has been added to your balance.')
            return 100
        else:
            print('Congratulations, you matched 3 basic numbers. $7 has been added to your balance.')
            return 7
    elif matches==2:
        if powerball:
            print('Congratulations, you matched 2 basic numbers and the powerball. $7 has been added to your balance.')
            return 7
        else:
            print('Congratulations, you matched 2 basic numbers. $4 has been added to your balance.')
            return 4
    elif matches==1:
        if powerball:
            print('Congratulations, you matched 1 basic number and the powerball. $4 has been added to your balance.')
            return 4
        else:
            print('Congratulations, you matched 1 basic number. $2 has been added to your balance.')
            return 2
    else:
        if powerball:
            print('Congratulations, you matched the powerball. $4 has been added to your balance.')
            return 4
        else:
            return 0

def display_results(total_tickets,current_balance):
    '''
    A program which displays the results from the 
    dictionary labeled results.
    '''
    print('\nFinal Results:\n')
    print(f'Total Tickets Bought:{total_tickets}')
    for items in results.values():
        print(items[0]+str(items[1]))
    print(f'Remaining funds:{current_balance}')  

results={
    (5,True):['Jackpots: ',0],
    (5,False):['Five Matches Only: ',0],
    (4,True):['Four Matches + Powerball: ',0],
    (4,False):['Four Matches Only: ',0],
    (3,True):['Three Matches + Powerball: ',0],
    (3,False):['Three Matches Only: ',0],
    (2,True):['Two Matches + Powerball: ',0],
    (2,False):['Two Matches Only: ',0],
    (1,True):['One Match + Powerball: ',0],
    (1,False):['One Match Only: ',0],
    (0,True):['Powerball Only: ',0],
    (0,False):['Losing Tickets: ',0]
}
mainframe()   
```
